# AbstractTensor — Manifolds & Indices

This notebook tests `indices.jl` and `manifolds.jl` step by step.

**Design reminder**
- A `TensorIndex` is a `(symbol, vbundle)` pair. The vbundle encodes variance completely.
- `up(:a1)` → lives in `:TangentM` (contravariant)
- `down(:a1)` → lives in `:CoTangentM` (covariant)
- `Up` / `Down` are syntax helpers only — not stored in the struct.
- `VBundleRecord.isdual` is the single authoritative source of bundle variance.

---
## 1. Loading package

```julia
using AbstractTensors
```

---
## 2. Define two manifolds

- **M** (dimension 4) with indices `a1, a2, a3, a4`
- **N** (dimension 4) with indices `b1, b2, b3, b4`

Each `@defManifold` call automatically creates:
- a tangent bundle (`:TangentM`, `isdual = false`)
- a cotangent bundle (`:CoTangentM`, `isdual = true`)
- registers all indices to the tangent bundle in `_IDX_REGISTRY`

```julia
@def_manifold M 4 [a1, a2, a3, a4]
```

```julia
@def_manifold N 4 [b1, b2, b3, b4]
```

---
## 3. Inspect the registries

After `@defManifold`, three registries are populated:
- `_MANIFOLDS` — one entry per manifold
- `_VBUNDLES`  — two entries per manifold (tangent + cotangent)
- `_IDX_REGISTRY` — one entry per index symbol → home tangent bundle

```julia
println("=== _MANIFOLDS ===")
for (k, v) in _MANIFOLDS
    println("  ", v)
end
```

```julia
println("=== _VBUNDLES ===")
for (k, v) in _VBUNDLES
    println("  $(v.name)  |  manifold=$(v.manifold)  |  dim=$(v.dim)  |  isdual=$(v.isdual)")
end
```

```julia
println("=== _IDX_REGISTRY ===")
for (sym, vb) in sort(collect(_IDX_REGISTRY), by = x -> string(x[1]))
    println("  :$sym  →  $vb")
end
```

---
## 4. Manifold accessors

```julia
println("is_manifold(M)         = ", is_manifold(M))
println("is_manifold(Float64)   = ", is_manifold(Float64))
println("dim_of_manifold(M)     = ", dim_of_manifold(M))
println("tangent_bundle_of(M)   = ", tangent_bundle_of(M))
println("cotangent_bundle_of(M) = ", cotangent_bundle_of(M))
println("vbundles_of(M)         = ", vbundles_of(M))
```

```julia
println("is_manifold(N)         = ", is_manifold(N))
println("dim_of_manifold(N)     = ", dim_of_manifold(N))
println("tangent_bundle_of(N)   = ", tangent_bundle_of(N))
println("cotangent_bundle_of(N) = ", cotangent_bundle_of(N))
```

---
## 5. Bundle variance via `isdual`

`isdual` is the single authoritative source — no naming convention is assumed.

```julia
println("_VBUNDLES[:TangentM].isdual    = ", _VBUNDLES[:TangentM].isdual)    # false
println("_VBUNDLES[:CoTangentM].isdual  = ", _VBUNDLES[:CoTangentM].isdual)  # true
println("_VBUNDLES[:TangentN].isdual    = ", _VBUNDLES[:TangentN].isdual)    # false
println("_VBUNDLES[:CoTangentN].isdual  = ", _VBUNDLES[:CoTangentN].isdual)  # true
```

```julia
println("is_tangent_bundle(:TangentM)     = ", is_tangent_bundle(:TangentM))    # true
println("is_tangent_bundle(:CoTangentM)   = ", is_tangent_bundle(:CoTangentM))  # false
println("is_cotangent_bundle(:CoTangentM) = ", is_cotangent_bundle(:CoTangentM))# true
println("is_cotangent_bundle(:TangentM)   = ", is_cotangent_bundle(:TangentM))  # false
```

---
## 6. `dual_bundle` — name-independent bundle duality

```julia
println("dual_bundle(:TangentM)    = ", dual_bundle(:TangentM))    # :CoTangentM
println("dual_bundle(:CoTangentM)  = ", dual_bundle(:CoTangentM))  # :TangentM
println("dual_bundle(:TangentN)    = ", dual_bundle(:TangentN))    # :CoTangentN
println("dual_bundle(:CoTangentN)  = ", dual_bundle(:CoTangentN))  # :TangentN
```

---
## 7. Index registration

```julia
# All a-indices registered to TangentM
for s in [:a1, :a2, :a3, :a4]
    println("index_registered(:$s) = $(index_registered(s))  |  home = $(index_home_vbundle(s))")
end
```

```julia
# All b-indices registered to TangentN
for s in [:b1, :b2, :b3, :b4]
    println("index_registered(:$s) = $(index_registered(s))  |  home = $(index_home_vbundle(s))")
end
```

```julia
# Unregistered index
println("index_registered(:z) = ", index_registered(:z))   # false
```

---
## 8. Constructing `TensorIndex` with `up` and `down`

`up` and `down` are **syntax-layer** helpers. They resolve the correct vbundle
from the registry at construction time. After that, only `(symbol, vbundle)` is stored.

```julia
println("up(:a1)   = ", up(:a1))    #  a1 ∈ TangentM
println("down(:a1) = ", down(:a1))  # -a1 ∈ CoTangentM
println()
println("up(:b1)   = ", up(:b1))    #  b1 ∈ TangentN
println("down(:b1) = ", down(:b1))  # -b1 ∈ CoTangentN
```

```julia
up(:a1)
```

---
## 9. Accessors on `TensorIndex`

```julia
println("symbol_of(up(:a2))  = ", symbol_of(up(:a2)))   # :a2
println("vbundle_of(up(:a2)) = ", vbundle_of(up(:a2)))  # :TangentM
println()
println("symbol_of(down(:a2))  = ", symbol_of(down(:a2)))   # :a2
println("vbundle_of(down(:a2)) = ", vbundle_of(down(:a2)))  # :CoTangentM
```

---
## 10. Variance predicates

```julia
println("is_contravariant(up(:a1))   = ", is_contravariant(up(:a1)))   # true
println("is_contravariant(down(:a1)) = ", is_contravariant(down(:a1))) # false
println("is_covariant(down(:a1))     = ", is_covariant(down(:a1)))     # true
println("is_covariant(up(:a1))       = ", is_covariant(up(:a1)))       # false
println()
println("is_up(up(:a3))    = ", is_up(up(:a3)))    # true
println("is_down(down(:a3)) = ", is_down(down(:a3))) # true
```

---
## 11. `flip` — moves an index to its dual bundle

```julia
println("flip(up(:a1))   = ", flip(up(:a1)))    # -a1 ∈ CoTangentM
println("flip(down(:a1)) = ", flip(down(:a1)))  #  a1 ∈ TangentM
println()
# flip is an involution
println("flip(flip(up(:a2))) == up(:a2)   = ", flip(flip(up(:a2))) == up(:a2))   # true
println("flip(flip(down(:a2))) == down(:a2) = ", flip(flip(down(:a2))) == down(:a2)) # true
```

---
## 12. `dual_vbundles` and `contractable`

Contraction requires: **same symbol** and **dual bundles of the same manifold**.

```julia
println("dual_vbundles(:TangentM, :CoTangentM) = ", dual_vbundles(:TangentM, :CoTangentM))  # true
println("dual_vbundles(:TangentM, :TangentM)   = ", dual_vbundles(:TangentM, :TangentM))    # false
println("dual_vbundles(:TangentM, :TangentN)   = ", dual_vbundles(:TangentM, :TangentN))    # false — different manifolds
println("dual_vbundles(:TangentM, :CoTangentN) = ", dual_vbundles(:TangentM, :CoTangentN))  # false — different manifolds
```

```julia
println("--- contractable ---")
println("up(:a1) vs down(:a1)  = ", contractable(up(:a1), down(:a1)))  # true  — same symbol, dual bundles
println("up(:a1) vs up(:a1)    = ", contractable(up(:a1), up(:a1)))    # false — same bundle
println("up(:a1) vs down(:a2)  = ", contractable(up(:a1), down(:a2)))  # false — different symbols
println("up(:a1) vs down(:b1)  = ", contractable(up(:a1), down(:b1)))  # false — different manifolds
println("up(:b1) vs down(:b1)  = ", contractable(up(:b1), down(:b1)))  # true  — same symbol, dual bundles on N
```

---
## 13. Equality and hashing

```julia
println("up(:a1) == up(:a1)   = ", up(:a1) == up(:a1))    # true
println("up(:a1) == down(:a1) = ", up(:a1) == down(:a1))  # false
println("up(:a1) == up(:a2)   = ", up(:a1) == up(:a2))    # false
println()
# hash consistency
println("hash stable: ", hash(up(:a1)) == hash(up(:a1)))  # true
```

## 15. `@indices` — add extra indices after definition

```julia
@add_indices M a5 a6
```

```julia
println("index_registered(:a5)      = ", index_registered(:a5))        # true
println("index_home_vbundle(:a5)    = ", index_home_vbundle(:a5))      # :TangentM
println("up(:a5)                    = ", up(:a5))                      #  a5 ∈ TangentM
println("down(:a6)                  = ", down(:a6))                    # -a6 ∈ CoTangentM
println("contractable(up(:a5), down(:a5)) = ", contractable(up(:a5), down(:a5)))  # true
```

---
## 16. `validate_contraction` — descriptive errors

```julia
# Valid contraction — should be silent
validate_contraction(up(:a1), down(:a1))
println("validate_contraction(up(:a1), down(:a1)) passed")
```

```julia
# Different symbols — should error
try
    validate_contraction(up(:a1), down(:a2))
catch e
    println("Caught: ", e.msg)
end
```

```julia
# Different manifolds — should error
try
    validate_contraction(up(:a1), down(:b1))
catch e
    println("Caught: ", e.msg)
end
```

---
## 17. Expected errors from unregistered symbols

```julia
# up on unregistered symbol
try
    up(:z)
catch e
    println("up(:z) → ", e.msg)
end
```

```julia
# dual_bundle on unregistered bundle
try
    dual_bundle(:TangentZ)
catch e
    println("dual_bundle(:TangentZ) → ", e.msg)
end
```

```julia
# @indices on unregistered manifold
try
    @indices Z c1
catch e
    println("@indices Z c1 → ", e.msg)
end
```

---
## 18. `@undefManifold` — clean teardown

```julia
@undefManifold N
```

```julia
println("haskey(_MANIFOLDS, :N)     = ", haskey(_MANIFOLDS, :N))       # false
println("haskey(_VBUNDLES, :TangentN)   = ", haskey(_VBUNDLES, :TangentN))    # false
println("haskey(_VBUNDLES, :CoTangentN) = ", haskey(_VBUNDLES, :CoTangentN))  # false
println("index_registered(:b1)      = ", index_registered(:b1))        # false
println("index_registered(:b4)      = ", index_registered(:b4))        # false
```

```julia
# M and its indices must be unaffected
println("is_manifold(M)           = ", is_manifold(M))              # true
println("index_registered(:a1)    = ", index_registered(:a1))       # true
println("index_home_vbundle(:a1)  = ", index_home_vbundle(:a1))     # :TangentM
```
